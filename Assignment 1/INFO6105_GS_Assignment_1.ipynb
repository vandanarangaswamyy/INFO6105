# Part 2: Visualizing the stock price history of Goldman Sachs from January 1, 2020 to 31st December 2024

```python
from datetime import datetime
import numpy as np
import scipy.stats as ss
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import expon, beta, norm
import seaborn as sns
from scipy.optimize import minimize
from pandas_datareader import DataReader

start_date = datetime(2020, 1, 1)
end_date = datetime(2024, 12, 31)

stock_symbol = "GS"
stock_df = yf.download(stock_symbol, start=start_date, end=end_date)
stock_df.index = pd.to_datetime(stock_df.index)

stock_df.head(10)
```

```python
# Visualization of stock for GS from 2020 - 2024

up_days = stock_df[stock_df['Close'] >= stock_df['Open']]
down_days = stock_df[stock_df['Close'] < stock_df['Open']]

plt.figure(figsize=(14, 7))
plt.plot(up_days.index, up_days['Low'], color='green', linestyle='-', marker='|')
plt.plot(up_days.index, up_days['High'], color='green', linestyle='-', marker='|')
plt.plot(down_days.index, down_days['Low'], color='red', linestyle='-', marker='|')
plt.plot(down_days.index, down_days['High'], color='red', linestyle='-', marker='|')

plt.title('Daily Stock High and Low for Goldman Sachs from January 2020 - December 2024')
plt.xlabel('Timeline')
plt.ylabel('Stock Price')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

# (1) the mixture modeling technique to approximate the empirical distribution using normal distributions as component distribution,





```python
stock_symbol = "GS"
X = stock_df['Close'].values
mu = np.mean(X)
se = np.std(X)

mu_1 = 180
sigma1 = 60
mu_2 = 320
sigma2 = 30
mu_3 = 430
sigma3 = 80
p1 = 0.30
p2 = 0.45
p3 = 1 - (p1 + p2)

T = 2000

r = np.zeros(T)
for t in range(T):
    r1 = mu_1 + sigma1 * np.random.normal(0,1,1)
    r2 = mu_2 + sigma2 * np.random.normal(0,1,1)
    r3 = mu_3 + sigma3 * np.random.normal(0,1,1)

    u = np.random.uniform(0,1,1)
    r[t] = r1 * (u <= p1)+ r2 * (u > p1)*(u <= p1 + p2) \
        + r3 * (u > p1 + p2 )*(u <= p1 + p2 + p3)

plt.figure(figsize=(10,6))
fig, ax = plt.subplots()
sns.kdeplot(data=X, linewidth=4)
plt.hist(r, bins=100, density=True, alpha=0.6, color='green', label="Histogram")
ax.legend([f'Empirical Kernal Distribution: {stock_symbol}', 'Mixture Model Simulated Distribution'])
```

# (2) bootstrap analysis to find the 95% confidence interval of the mean estimates

```python
T = X.shape[0]
B = 1000
mu_boot = np.zeros(B)
se_boot = np.zeros(B)
x_boot_std = np.zeros(B)

for i in range(B):
    x_boot = X[np.random.choice(T, T)]
    mu_boot[i] = np.mean(x_boot)
    se_boot[i] = np.std(x_boot) / np.sqrt(T)
    x_boot_std[i] = np.std(x_boot)

mu_boot = np.sort(mu_boot)
se_boot = np.sort(se_boot)
x_boot_std = np.sort(x_boot_std)

print("Statistics about the Mean")
print("95% confidence interval of mu_boot:", mu_boot[25], mu_boot[975])
print("95% confidence interval of standard error of mu:", se_boot[25], se_boot[975])
print("\n")

print("Statistics about the Sample")
print(f"95% confidence interval of {stock_symbol} stock price std (method 1):", se_boot[25] * np.sqrt(T), se_boot[975] * np.sqrt(T))
print(f"95% confidence interval of {stock_symbol} stock price std (method 2):", x_boot_std[25], x_boot_std[975])
```

# (3) Monte-Carlo analysis to simulate 1000 paths using five key parameters that mimic the empirical distribution of your stock.

```python
mu_1 = 188
sigma1 = 42
mu_2 = 320
sigma2 = 20
mu_3 = 445
sigma3 = 80
p1 = 0.30
p2 = 0.45
p3 = 1 - (p1 + p2)

def generateEmpiricalDist(mu_1, sigma1, mu_2, sigma2, mu_3, sigma3, p1, p2, p3, T):
    simulatedPrice = np.zeros(T)

    for t in range(T):
        u = np.random.uniform(0, 1)

        if u <= p1:
            simulatedPrice[t] = np.random.normal(mu_1, sigma1)
        elif u <= p1 + p2:
            simulatedPrice[t] = np.random.normal(mu_2, sigma2)
        else:
            simulatedPrice[t] = np.random.normal(mu_3, sigma3)

    return simulatedPrice


T = X.shape[0]
M = 1000
mu_mc = np.zeros(M)
y_mc = 0
for i in range(0, M):
    simulatedPrice = generateEmpiricalDist(mu_1, sigma1,\
                                           mu_2, sigma2,\
                                           mu_3, sigma3,\
                                           p1, p2, p3, T)
    mu_mc[i] = np.mean(simulatedPrice)
    y_mc += np.sort(simulatedPrice)

y_mc = y_mc/M
mu_mc = np.sort(mu_mc)

print("confidence interval of mu_mc:", mu_mc[25], mu_mc[975])

# Visualization
fig, ax = plt.subplots(figsize=(10,6))
sns.kdeplot(data=X, linewidth=4)
sns.kdeplot(data=y_mc, linewidth=4)
ax.legend(['Empirical Kernal Distribution', 'Monte-Carlo Simulated Distribution'], loc = 'upper right')
```

```python
stock_df['Log Returns'] = np.log(stock_df['Close'] / stock_df['Close'].shift(1))
stock_df = stock_df.dropna()

S0 = float(stock_df['Close'].iloc[-1])
T = 252
N = 1000

mu = np.mean(stock_df['Log Returns'])
sigma = np.std(stock_df['Log Returns'])

print(f"Initial stock price: {S0:.2f}")
print(f"Mean log return: {mu:.6f}")
print(f"Standard deviation (volatility): {sigma:.6f}")

def generate_stock_path(T, mu, sigma, S0):
    dt = 1 / T
    price_path = np.zeros(T)
    price_path[0] = S0

    for t in range(1, T):
        Z = np.random.normal(0, 1)
        price_path[t] = price_path[t-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)

    return price_path

simulated_paths = np.zeros((T, N))
for i in range(N):
    simulated_paths[:, i] = generate_stock_path(T, mu, sigma, S0)

plt.figure(figsize=(12, 6))
plt.plot(simulated_paths, color="blue", alpha=0.2)
plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.title(f"Monte Carlo Simulation of GS Stock ({N} Paths)")
plt.show()
```
